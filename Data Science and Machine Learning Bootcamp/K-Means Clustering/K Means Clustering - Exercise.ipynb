# K Means Clustering - Exercise

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8,
                 random_state=101)
```

```python
plt.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow', edgecolors='black', s=25)
```

```python
data[1]
```

## K Means Clustering

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.labels_
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, sharey=True, figsize=(10,6))

ax1.set_title('K Means')
ax1.scatter(data[0][:,0], data[0][:,1], c=kmeans.labels_, cmap='coolwarm', s=25, edgecolors='black')

ax2.set_title('Original')
ax2.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='plasma', s=25, edgecolors='black')
```
